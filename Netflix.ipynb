It is important to note that the Netflix dataset contains missing values in several columns, indicating a lack of available information for those specific data points. These missing values can potentially impact the analysis and interpretations derived from the dataset. It is recommended to handle these missing values by employing appropriate data preprocessing techniques before conducting any analysis. By addressing the missing values, such as through imputation or dropping rows, the dataset can be made more complete and reliable for accurate analysis and insights. However, if the decision is made to not modify or alter the dataset in any way, it is crucial to acknowledge that the missing values will remain, and any subsequent analysis should take into consideration the potential limitations and uncertainties arising from these missing values.

```python
df['director'].fillna('No Data Available', inplace=True)
df['cast'].fillna('No Data Available', inplace=True)
df['country'].fillna('No Data Available', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
# Fill missing values in numeric columns with mean or custom values
df['release_year'] = df['release_year'].fillna(df['release_year'].mean())
df['duration'] = df['duration'].fillna('Unknown')
```

```python
df.dropna(inplace=True)
```

# Data Visulization

## **1.Which country has produced the highest number of movies and TV shows in the dataset? Are there any notable variations in content production across different countries?**

```python
# Count the occurrences of each country in the 'country' column
country_counts = df['country'].value_counts()

# Get the top 10 countries with the highest content production
top_countries = country_counts.head(10)

colors = plt.cm.tab20c(range(len(top_countries)))

# Plot the content production by country with color
plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar', color=colors)
plt.xlabel('Country')
plt.ylabel('Number of Movies and TV Shows')
plt.title('Content Production by Country')
plt.xticks(rotation=45)
plt.show()
```

# 2.How has the number of TV shows and movies in the Netflix dataset changed over the years? Are there any noticeable trends or patterns?

```python
# Create a new DataFrame with the release year and type columns
df_year_type = df[['release_year', 'type']]

# Group the data by release year and type and count the occurrences
count_year_type = df_year_type.groupby(['release_year', 'type']).size().unstack()

# Plot the number of TV shows and movies over the years
plt.figure(figsize=(12, 8))
count_year_type.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Number of TV Shows and Movies over the Years')
plt.legend(['Movie', 'TV Show'])
plt.grid(True)
plt.show()
```

# 3.What are the most popular genres among Netflix users? How has the popularity of different genres evolved over time?

```python
# Extract the genres from the 'listed_in' column
df['genres'] = df['listed_in'].apply(lambda x: x.split(','))

# Create a new DataFrame with the exploded genres
df_exploded = df.explode('genres')

# Group the data by genre and count the occurrences
genre_counts = df_exploded['genres'].value_counts()

# Select the top N genres with the highest counts
top_genres = genre_counts.head(10)  # Adjust N to your desired number of top genres

# Sort the genres in descending order of popularity
top_genres = top_genres.sort_values(ascending=True)

# Plot the popularity of genres over time
plt.figure(figsize=(10, 6))
top_genres.plot(kind='barh', color=['#FF5A5F', '#FFB400', '#3DADFF', '#00CC96', '#D4AEFF',
                                   '#8D44AD', '#FF7F40', '#FFD700', '#FF69B4', '#00BFFF'])
plt.xlabel('Count')
plt.ylabel('Genre')
plt.title('Popularity of Genres on Netflix')
plt.grid(True)
plt.show()
```

# 4.Which countries have experienced the highest growth in terms of content production on Netflix? Are there any emerging markets worth exploring for further expansion?

```python
# Filter the DataFrame to include only rows with non-null country values
df_filtered = df.dropna(subset=['country'])

# Count the occurrences of each country and type (TV show or movie) in the DataFrame
country_type_counts = df_filtered.groupby(['country', 'type']).size().unstack()

# Sort the countries based on the total content production
country_type_counts['Total'] = country_type_counts.sum(axis=1)
country_type_counts = country_type_counts.sort_values(by='Total', ascending=False)

# Select the top N countries with the highest total content production
top_countries = country_type_counts.head(10)  # Adjust N to your desired number of top countries

# Plot the growth in content production by country using a stacked bar graph
plt.figure(figsize=(15, 8))
top_countries.drop(columns='Total').plot(kind='barh', stacked=True)
plt.xlabel('Count')
plt.ylabel('Country')
plt.title('Growth in Content Production on Netflix by Country')
plt.legend(title='Type', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

# 5.Which rating has the highest among all in this dataset ?

```python
rating_counts = df.groupby('rating')['show_id'].count()
rating_counts
```

```python
# Remove the first three rows from the rating distribution
rating_counts = df['rating'].value_counts().drop(['66 min', '74 min', '84 min'])
plt.figure(figsize=(15, 6))
colors = plt.cm.Set3(np.linspace(0, 1, 14))
bar_plot = rating_counts.plot(kind='bar', color=colors)
bar_plot.set_xlabel("Rating")
bar_plot.set_ylabel("Count")
bar_plot.set_title("Count of Ratings")
for i, v in enumerate(rating_counts):
    bar_plot.text(i, v, str(v), ha='center', va='bottom')
plt.show()
```
